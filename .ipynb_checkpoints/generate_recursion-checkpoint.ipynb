```python
from igraph import plot
from igraph import Graph
import igraph
import matplotlib.pyplot as plt
```

```python
def fib_with_memory(n, parent_vertex=None) -> int:
    global g, memory, counter
    counter += 1
    name = str(n)
    current_vertex = g.add_vertex(name)
    current_vertex['label'] = name
    if parent_vertex is not None:
        g.add_edge(parent_vertex, current_vertex)
    if n == 0:
        current_vertex['label'] = 0
        return 0
    if n == 1 or n == 2:
        current_vertex['label'] = 1
        return 1
    if n in memory:
        current_vertex['label'] = memory[n]
        current_vertex['color'] = 'green'
        return memory[n]
    
    result = fib_with_memory(n-1, current_vertex) + fib_with_memory(n-2, current_vertex)
    current_vertex['label'] = result
    memory[n] = result
    return result

def fib(n, parent_vertex=None) -> int:
    global g, counter
    name = str(n)
    counter += 1
    current_vertex = g.add_vertex(name)
    current_vertex['label'] = name
    if parent_vertex is not None:
        g.add_edge(parent_vertex, current_vertex)
    if n == 0:
        current_vertex['label'] = 0
        current_vertex['color'] = 'white'
        return 0
    if n == 1:
        current_vertex['label'] = 1
        current_vertex['color'] = 'white'
        return 1
    
    result = fib(n-1, current_vertex) + fib(n-2, current_vertex)
    current_vertex['label'] = result
    return result

g = Graph(directed=True)
memory = dict()
counter = 0

fib_with_memory(8)

fig, ax = plt.subplots(figsize=(21,21)) # figsize=(14,14))
plot(
    g,
    bbox=(1000, 1000),
    vertex_label_size = 30,
    vertex_size = 80,
    target=ax,
    layout="rt"
)
print('count of calls fib = ', counter)
# plt.savefig("fib_4.jpg")
plt.show()
```
